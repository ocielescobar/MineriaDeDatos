# Caso 1: Uso de Transporte Público en Santiago

Este cuaderno muestra un ejemplo simple de **Naive Bayes** aplicado a un problema de clasificación.

## Contexto
La Municipalidad quiere predecir si una persona usará **Transporte Público** o **Auto Particular**, considerando variables como:

- distancia al trabajo
- nivel de ingreso
- clima
- hora del día

El objetivo es construir un modelo que clasifique el tipo de transporte esperado.

## 1. Importar librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, ConfusionMatrixDisplay
```

## 2. Crear el set de datos

Cada fila representa una persona con ciertas características y el medio de transporte que utiliza.

```python
data = [
    [2,  'Bajo',  'Soleado', 'Punta',  'Publico'],
    [10, 'Alto',  'Soleado', 'Normal', 'Auto'],
    [5,  'Medio', 'Lluvia',  'Punta',  'Publico'],
    [8,  'Alto',  'Lluvia',  'Normal', 'Auto'],
    [3,  'Bajo',  'Lluvia',  'Punta',  'Publico'],
    [12, 'Alto',  'Soleado', 'Punta',  'Auto'],
    [4,  'Medio', 'Soleado', 'Normal', 'Publico'],
    [7,  'Medio', 'Lluvia',  'Punta',  'Publico'],
    [15, 'Alto',  'Soleado', 'Normal', 'Auto'],
    [6,  'Bajo',  'Lluvia',  'Normal', 'Publico'],
    [9,  'Alto',  'Lluvia',  'Punta',  'Auto'],
    [1,  'Bajo',  'Soleado', 'Punta',  'Publico'],
    [11, 'Alto',  'Soleado', 'Normal', 'Auto'],
    [5,  'Medio', 'Soleado', 'Punta',  'Publico'],
    [13, 'Alto',  'Lluvia',  'Normal', 'Auto']
]

df = pd.DataFrame(data, columns=['distancia_km', 'ingreso', 'clima', 'hora', 'transporte'])
df
```

## 3. Revisar el dataset

```python
print('Cantidad de registros:', df.shape[0])
print('Cantidad de columnas:', df.shape[1])
print('\nTipos de datos:')
print(df.dtypes)
print('\nDistribución de la variable objetivo:')
print(df['transporte'].value_counts())
```

## 4. Gráfico 1: Distribución del tipo de transporte

```python
conteo_transporte = df['transporte'].value_counts()

plt.figure(figsize=(7,5))
plt.bar(conteo_transporte.index, conteo_transporte.values)
plt.title('Cantidad de personas por tipo de transporte')
plt.xlabel('Tipo de transporte')
plt.ylabel('Cantidad')
plt.show()
```

## 5. Gráfico 2: Distancia al trabajo según tipo de transporte

```python
promedios = df.groupby('transporte')['distancia_km'].mean()

plt.figure(figsize=(7,5))
plt.bar(promedios.index, promedios.values)
plt.title('Distancia promedio según tipo de transporte')
plt.xlabel('Tipo de transporte')
plt.ylabel('Distancia promedio (km)')
plt.show()
```

## 6. Codificar variables categóricas

Naive Bayes necesita trabajar con valores numéricos, por eso transformamos las columnas categóricas.

```python
encoders = {}

for col in ['ingreso', 'clima', 'hora', 'transporte']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le

df
```

## 7. Separar variables predictoras y variable objetivo

```python
X = df.drop('transporte', axis=1)
y = df['transporte']

print('X:')
print(X.head())
print('\ny:')
print(y.head())
```

## 8. Dividir en entrenamiento y prueba

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print('Tamaño entrenamiento:', X_train.shape)
print('Tamaño prueba:', X_test.shape)
```

## 9. Entrenar el modelo Naive Bayes

```python
modelo = GaussianNB()
modelo.fit(X_train, y_train)
```

## 10. Realizar predicciones

```python
y_pred = modelo.predict(X_test)

print('Valores reales:     ', list(y_test))
print('Valores predichos:  ', list(y_pred))
```

## 11. Evaluar el modelo

```python
matriz = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

print('Matriz de confusión:')
print(matriz)
print('\nAccuracy del modelo:', round(accuracy, 4))
print('\nReporte de clasificación:')
print(classification_report(y_test, y_pred))
```

## 12. Gráfico 3: Matriz de confusión

```python
disp = ConfusionMatrixDisplay(confusion_matrix=matriz, display_labels=encoders['transporte'].classes_)
disp.plot()
plt.title('Matriz de confusión del modelo')
plt.show()
```

## 13. Probar con un nuevo caso

Caso nuevo:

- distancia = 5 km
- ingreso = Medio
- clima = Lluvia
- hora = Punta

```python
nuevo_caso = pd.DataFrame([[5, 'Medio', 'Lluvia', 'Punta']], columns=['distancia_km', 'ingreso', 'clima', 'hora'])
nuevo_caso
```

```python
nuevo_caso_codificado = nuevo_caso.copy()

for col in ['ingreso', 'clima', 'hora']:
    nuevo_caso_codificado[col] = encoders[col].transform(nuevo_caso_codificado[col])

nuevo_caso_codificado
```

```python
prediccion = modelo.predict(nuevo_caso_codificado)
resultado = encoders['transporte'].inverse_transform(prediccion)

print('Predicción del modelo:', resultado[0])
```

## 14. Conclusión

En este caso, Naive Bayes permite estimar si una persona utilizará **transporte público** o **auto particular** a partir de variables simples.
